```python
# Calculate IQR
Q1 = df_cleaned.quantile(0.25)
Q3 = df_cleaned.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
# remove outlier
df_cleaned = df_cleaned[~((df_cleaned < (lower)) |(df_cleaned > (upper))).any(axis=1)]
```
