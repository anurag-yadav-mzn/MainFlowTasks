```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/householdtask3.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
# scatter plot
plt.scatter(df['year'],df['own'])
plt.title('Year against own')
plt.xlabel('Year')
plt.ylabel('Own')
plt.show()
```

```python
# Line chart
plt.plot(df['year'])
plt.plot(df['own'])
plt.title('Year against own')
plt.xlabel('Year')
plt.ylabel('Own')
plt.show()
```

```python
# bar plot
plt.bar(df['year'],df['own'])
plt.title('Year against own')
plt.xlabel('Year')
plt.ylabel('Own')
plt.show()
```

```python
# histogram
plt.hist(df['income'])
plt.title('Histogram of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```
